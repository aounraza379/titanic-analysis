```
import pandas as pd

# Load Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print("Dataset loaded successfully!")
```

```
# first 5 rows
print("HEAD (first 5 rows):")
print(df.head())

# last 5 rows
print("\nTAIL (last 5 rows):")
print(df.tail())

# Dataset dimensions (rows, columns)
print("\nSHAPE:", df.shape)

# Column names
print("\nCOLUMNS:", df.columns.tolist())

# Data types and non-null counts
print("\nINFO:")
df.info()

# Statistical summary
print("\nDESCRIBE:")
print(df.describe())
```

```
print('Missing Values: ')
print(df.isnull().sum())
```

```
# Check missing values per column
print("MISSING VALUES:")
print(df.isnull().sum())

# Fill missing Age with the median age (better than mean — not affected by outliers)
df['Age'].fillna(df['Age'].median(), inplace=True)

# Fill missing Embarked with the most common value (mode)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Drop the Cabin column — too many missing values to be useful
df.drop(columns=['Cabin'], inplace=True)

# Confirm no more missing values
print("\nMISSING VALUES AFTER CLEANING:")
print(df.isnull().sum())
```

```
# FILTERING — only passengers who survived
survivors = df[df['Survived'] == 1]
print("Total Survivors:", len(survivors))

# SORTING — sort by Age (youngest to oldest)
sorted_df = df.sort_values(by='Age', ascending=True)
print("\nYoungest 5 passengers:")
print(sorted_df[['Name', 'Age', 'Survived']].head())

# GROUPING — average fare and age per passenger class
grouped = df.groupby('Pclass')[['Age', 'Fare']].mean()
print("\nAverage Age and Fare by Passenger Class:")
print(grouped)

# SAVE the cleaned dataset
df.to_csv('titanic_cleaned.csv', index=False)
print("\nCleaned dataset saved as titanic_cleaned.csv")
```
